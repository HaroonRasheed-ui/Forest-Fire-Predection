```python
# Load the dataset
data = pd.read_csv('forestfires.csv')

# Convert month and day columns to numerical representations
months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
days = ['mon', 'tue', 'wed', 'thu', 'fri', 'sat', 'sun']

data['month'] = data['month'].apply(lambda x: months.index(x))
data['day'] = data['day'].apply(lambda x: days.index(x))

# Correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Distribution of the area (target variable)
sns.histplot(data['area'], bins=30, kde=True)
plt.show()
```
